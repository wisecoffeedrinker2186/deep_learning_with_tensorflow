#### Shuffle order of elements in a tensor


```python
# shuffle a tensor (helps shuffle the data so inherent position does not affect tyhe learning)
not_shuffled = tf.constant([[1,2],[3,4],[5,6]])
not_shuffled.ndim

# shufling the tensor (along 1st dimension)
tf.random.shuffle(not_shuffled)
```

```python
# shufling the tensor (along 1st dimension)
tf.random.set_seed(23)   # global level random seed
tf.random.shuffle(not_shuffled,seed=23) # local level random seed
```

```python
not_shuffled
```

Shuffled tensors to be in the same order use both global and local level random seed.

Matrix multiplication on tensors of different shapes

```python
# Create a tensor (3,2) tensor
X = tf.random.Generator.from_seed(23)
X = X.normal(shape=(3,2))

# Another tensor of same shape
Y = tf.random.Generator.from_seed(32)
Y = Y.normal(shape=(3,2))
```

```python
# tf.matmul(X,Y) -- this does not work due to incompatable shapes inner dimesions must match
```

Solution:

```python
# change the shape of y
tf.reshape(Y, shape=(2,3))
Y
```

```python
# now X with  reshaped Y
tf.matmul(X, tf.reshape(Y, shape=(2,3)))
```

OR

```python
# use transpose of the tensor
tf.matmul(X, tf.transpose(Y))
```

```python
tf.matmul(tf.transpose(X),Y)
```

**The dot product**

Matrix multiplication is referred to as dot product this can be done using ;-
* `tf.matmul`
* `tf.tensordot`

```python
# perfrom dot product on X and Y
tf.tensordot(tf.transpose(X),Y,axes=1)
```

```python
# do matrix multiplication with both transpose and reshape
tf.tensordot(X, tf.transpose(Y), axes=1) , tf.tensordot(X,tf.reshape(Y,shape=(2,3)),axes=1)
```

```python
# Check values of Y, transposed Y, reshaped Y
print('Normal Y:')
print(Y, '\n')

print('Y reshaped:') # here the values are shuffled to fit shape
print(tf.reshape(Y,shape=(2,3)), '\n')

print('Y transposed:') # here axis is swapped not shuffling
print(tf.transpose(Y), '\n')
```

Generally, when performing matrix multiplication on two tensors you will usually transpose one of the two tensors to satisfy the rules of matrix multiplication

Other forms of aggregation:

**Find the positional maximum and minimum of a tensor**

(helps fond the max probability in the output of the NN -- one hot encode the out)

```python
# creating a random tensor
tf.random.set_seed(23)
F = tf.random.uniform(shape=[50])
F
```

```python
# finding positional maximum
tf.argmax(F)
```

```python
# index on our largest value position
F[tf.argmax(F)]
```

```python
#Confirm the largest value in the F tensor
tf.reduce_max(F)
```

```python
# confirming the max
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
# positional minimum
tf.argmin(F)
```

```python
# finding minimum using positional index
F[tf.argmin(F)]
```
